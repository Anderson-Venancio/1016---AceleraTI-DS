```python
import numpy as np

#Primeira coluna é a taxa de compra e a segunda a taxa de venda de NTN-Bs 2035 no ano de 2023, série principal.
#Nem todas as operações fazem sentido, são apenas demonstrações de conceito usando numpy

data = np.genfromtxt('NTN-B_Principal_2023.csv',  delimiter=';', dtype=float)

data = np.nan_to_num(data, nan=6.14)

mean_tx = np.mean(data, axis=0)
min_tx = np.min(data, axis=1)
max_tx = np.max(data)

print(f'Média das taxas de compra coluna 1: {mean_tx[0]}')
print(f'Média das taxas de venda coluna 2: {mean_tx[1]}')

print(f'Taxa mínima de cada dia: {min_tx}') #Não faz sentido para estes dados, apenas conceito

print(f'Maior taxa do período: {max_tx}') #Não faz sentido para estes dados, apenas conceito

#diferença entre a taxa mais recente (fim de julho), e as taxas do início do ano, usando transposta
dif_tx_ini_fim_periodo = data.T[:, 0] - data.T[:, len(data)-1]
print(f'Diferença entre as taxas de compra e de venda do início e do fim do período (transposta): {dif_tx_ini_fim_periodo}')

#A conta não faz sentido, é apenas para demonstrar o conceito do produto escalar
produto_escalar = data[:, 0] @ data[:, 1]
print(f'Produto Escalar: {produto_escalar}')

# Filtro para taxas acima de 6.5
filtro = data > 6.5
print(f'Filtro, taxas > 6.5: {data[filtro]}')

print(f'Matrix * número, dobrando as taxas de compra e triplicando as de venda: {data * [2,3]}')
```
